```python
df.columns
```

```python
x = sns.countplot(x = "Gender",data = df)
for container in x.containers:
    x.bar_label(container)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
ab = sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

***From abobe Graphs we can see that most of the buyers are females and even purchasing power are greater than male***

## Age

```python
df.columns
```

```python
ab = sns.countplot(data = df , x = 'Age Group', hue = 'Gender')
for container in ab.containers:
    ab.bar_label(container)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = "Amount",ascending = False)
xy = sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

***From above graphs we can see that most of the buyers are the age group between 26-35 yrs female***

## State

```python
df.columns
```

```python
# Total number of order from top 10 states
state_ord = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(x = 'State' ,y = 'Orders' , data = state_ord)
```

```python
# Total amount/sales from top 10 states
sales_amt = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = "Amount" ,ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (20,5),'font.size' : 15})
x = sns.barplot(x = 'State' , y = 'Amount' , data = sales_amt)
```

***From above graphs we can see that most of the orders and total sales/amount are from Uttar Pradesh,Maharashtra and 
Karnatka respectively***

## Marital Status

```python
df.columns
```

```python
ex = sns.countplot(x = 'Marital_Status',data = df)
for container in ex.containers:
    ex.bar_label(container)
sns.set(rc = {'figure.figsize' : (4,5)})
```

```python
marital_sales = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
ab = sns.barplot(x = 'Marital_Status',y = 'Amount',data = marital_sales,hue = 'Gender')
```

***From above graphs we can see that most of the buyers are married (women) and they have high purchasing power***

## Occupation

```python
df.columns
```

```python
ab = sns.countplot(x = 'Occupation',data = df)
sns.set(rc = {'figure.figsize' : (1,5)})
for container in ab.containers:
    ab.bar_label(container)
```

```python
a = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Occupation',y = 'Amount',data = a)
```

***From above graphs we can see that most of the buyers are working in IT,Avation and Healthcare sector***

## Product Category

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize': (25,10)})
n = sns.countplot(x = 'Product_Category',data = df)
for container in n.containers:
    n.bar_label(container)
```

```python
mc = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.barplot(x = 'Product_Category',y = 'Amount',data = mc)
sns.set(rc = {'figure.figsize': (25,5)})
```

***From above graphs we can see that most of the sold products are from food,clothing and electronics.***

```python
df.columns
```

```python
mk = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.barplot(x= 'Product_ID',y = 'Orders',data = mk)
sns.set(rc = {'figure.figsize': (15,5)})
```

```python
# Top 10 most sold products(same thing as above)
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending = False).plot(kind = 'bar')
```

## Conclusion:

***Married women age group 26-35yrs from UP,Maharashtra and Karnatka working in IT,healthcare and aviation are more likely to buy products from food,clothing and electronics category.***
